```python
import pandas
import pandas as pd
```

```python


Revenue_Growth_Factor = 1.18
EBITDA_Margin = 0.45
Taxes = 0.21
Capital_Expenditures = 0.3
Working_Capital_Increase = 0.06
EBITDA_Exit_Multiple = 32.70
Discount_Factor = 0.94
```

```python
dates=pandas.date_range(start='2022',end='2032',freq='Y')
dates
```

```python
model_df=pandas.DataFrame(index=dates)
terminal_value_df=pandas.DataFrame()
model_df.T
```

```python
statrt_Revenue = 222
rev_projection= [statrt_Revenue * Revenue_Growth_Factor]
for i in range (1,len(dates)):
   rev_projection.append(rev_projection[i-1] * Revenue_Growth_Factor)
model_df['Revenue']= rev_projection
model_df[['Revenue']].T
```

```python
model_df['EBITDA']= model_df['Revenue'] * EBITDA_Margin
model_df.T
```

```python
model_df['Taxes']= model_df['EBITDA'] * -Taxes
model_df.T
```

```python
model_df['Capital Expenditures']= model_df['Revenue'] * -Capital_Expenditures
model_df.T
```

```python
model_df['Working Capital Increase']= model_df['Revenue'] * -Working_Capital_Increase
model_df.T.round(2)
```

```python
model_df['Cash_flow']= model_df.drop(columns=['Revenue']).sum(axis=1)
model_df.T.round(0)
```

```python
Discount_Factor=[Discount_Factor ** i for i in range(1,len(dates)+1)]
dis_df = pandas.DataFrame(index = Discount_Factor)
dis_df['fcf value'] = Discount_Factor * model_df['Cash_flow'].values
dis_df.T
```

```python
present_value_of_projected_fcf=dis_df['fcf value'].sum()
present_value_of_projected_fcf
```

```python
final_df=pandas.DataFrame(index=['2030'])
final_df['Revenue']=model_df['Revenue'].iloc[-1]*Revenue_Growth_Factor
final_df['EBITDA']= final_df['Revenue']* EBITDA_Margin
```

```python
final_df['final_valuation']= final_df['EBITDA']* EBITDA_Exit_Multiple
final_df['final_valuation'].round(0)
```
